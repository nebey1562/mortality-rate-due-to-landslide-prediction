```python
import pandas as pd
```

```python
df=pd.read_csv("lanslide_global.csv")
```

```python
df=df.astype(str)
df
```

```python
df.columns
```

```python
df=df[['event_description', 'location_description',
       'location_accuracy','landslide_trigger',
       'landslide_size', 'landslide_setting', 'landslide_category','fatality_count', 'injury_count',
       'storm_name', 'photo_link', 'notes', 'event_import_source',
       'event_import_id', 'country_name', 'country_code','admin_division_population']]
df
```

```python
#trigger
df["rain"]=df['landslide_trigger'].apply(lambda x: 1 if ('rain' in x.lower()) or ('monsoon' in x.lower()) or ('flood' in x .lower()) or ('downpour' in x.lower()) else 0)
df['earthquake']=df['landslide_trigger'].apply(lambda x: 1 if "earthquake" in x.lower() else 0)
df["downpour"]=df['landslide_trigger'].apply(lambda x: 1 if  ('flood' in x.lower()) or ('downpour' in x.lower()) else 0)
df['snow']=df['landslide_trigger'].apply(lambda x: 1 if ('snow' in x.lower()) or ('freeze' in x.lower()) else 0)
df['other']=df['landslide_trigger'].apply(lambda x: 1 if ('other'in x.lower()) or ('cyclone' in x.lower()) or ('no_apparent_trigger' in x) or ('volcanic' in x.lower()) or ('construction' in x.lower()) else 0)
```

```python
#scale
df['small_scale']=df['landslide_size'].apply(lambda x: 1 if "samll" in x.lower() else 0)
df['medium_scale']=df['landslide_size'].apply(lambda x: 1 if "medium" in x.lower() else 0)
df['large_scale']=df['landslide_size'].apply(lambda x: 1 if "large" in x.lower() else 0)
```

```python
df.downpour.value_counts()
```

```python
df.rain.value_counts()
```

```python
#remove nan from population divsion
df['admin_division_population']=df['admin_division_population'].apply(lambda x: 'nan' if '0.0' in x.lower() else x)
```

```python
df
```

```python
df.columns
```

```python
df['home_destroyed']=df['event_description'].apply(lambda x: 1 if 'home' in x.lower() else 0)
```

```python
df['home_destroyed'].value_counts()
```

```python
df['mudslide']=df['landslide_category'].apply(lambda x: 1 if 'mudslide' in x.lower() else 0)
df['landslide']=df['landslide_category'].apply(lambda x: 1 if 'landslide' in x.lower() else 0)
```

```python
df
```

```python
if df['fatality_count'].empty and df['injury_count'].empty:
    df=df.dropna()
df
```

```python
df['injury_count']=df['injury_count'].apply(lambda x: '0.0' if 'nan' in x.lower() else x)
df['fatality_count']=df['fatality_count'].apply(lambda x: '0.0' if 'nan' in x.lower() else x)
```

```python
df
```

```python

df1 = df.loc[df["admin_division_population"] !='nan' ]
condition = (df1['injury_count'] == '0.0') & (df1['fatality_count'] == '0.0')
df1.drop(df1[condition].index, inplace=True)

df1
```
